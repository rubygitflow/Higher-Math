Задача 6

Три игрока сели играть в покер. Сколькими способами им можно сдать по одной карте? (колода содержит 36 карт)

```
# Для начала вычислим, сколькими способами можно извлечь 3 карты из колоды (по оджной трем игрокам)
c_3 = C_n_m(36,3)
print("{} способами можно извлечь 3 карты".format(c_3))

# рассмотрим одну из 7140 возможных комбинаций, например: Король пик, 6 червей , 5 треф. 
# эти 3 карты можно распределить между игроками следующими способами:

# КП, 6Ч, 5Т;
# КП, 5Т, 6Ч;
# 6Ч, КП, 5Т;
# 6Ч, 5Т, КП;
# 5Т, КП, 6Ч;
# 5Т, 6Ч, КП.

P = factorial(3)
print("{} способами можно распределить 3 карты извлеченные ихз колоды".format(P))

A =  A_(36,3)
print("{} способами можно сдать по одной карте участникам".format(A))
```

Задача 7

В фирме работает 23 человек, среди которых 10 мужчин и 13 девушек. Сколькими способами можно выбрать двух человек одинакового пола?

```
# Мы не можем сразу воспользоваться формулой C_n_m поскольку общее количество сочетаний включает в себя пары разного пола.
# Условие подразумевает, что необходимо выбрать ИЛИ! двух мужчин ИЛИ! двух девушек
m = C_n_m(10,2)
print("{} способами можно выбрать 2 мужчин".format(m))

d = C_n_m(13,2)
print("{} способами можно выбрать 2 девушек".format(d))

res = m + d
print("{} способами можно выбрать двух человек одинакового пола?".format(res))
```
