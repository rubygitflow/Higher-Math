```
# Способ 1 (самый быстрый для графиков, но не знаем формулы производной)
curr_time = time.time()
x = np.linspace(-10,10,2001)
dx = x[1]-x[0]
y = x**2 + 1
dydx = np.gradient(y, dx)
print(f"Время расчёта: {time.time() - curr_time}")

plt.plot(y, label=f'x**2 + 1')
plt.plot(dydx, label=' np.gradient')
plt.legend()
plt.show()
print (f"Производная функции x**2 + 2 = {dydx}")
```

```
# Способ 2 (самый медленный для графиков, но получили на вывод формулу производной)
curr_time = time.time()
drange = np.linspace(-10,10,2001)
x = sp.Symbol('x')
y = x**2 + 1
dydx = y.diff(x)
_y = np.array([y.evalf(subs={x:i}) for i in drange])
_dydx = np.array([dydx.evalf(subs={x:i}) for i in drange])
print(f"Время расчёта: {time.time() - curr_time}")

plt.plot(_y, label=f'{y}')
plt.plot(_dydx, label=f'{dydx}')
plt.legend()
plt.show()
print (f"Производная функции {y} = {dydx}")
```

```
# Способ 3 (не быстрый для выавода графиков, и программировали функции)
#Функция вычисления производной функции
def get_deriviate(f, drange):
  #Вычисляем координаты массива x
  x = drange
  len_x = x.shape[0]

  #Рассчитываем производную
  dydx = np.array([(f(x[i+1]) - f(x[i]))/(x[i+1]-x[i]) for i in range(len_x-1)])
  #Возвращаем производную
  return dydx

curr_time = time.time()
x = np.linspace(-10,10,2000)
_x = x**2+1
_dydx = get_deriviate(lambda x: x**2+1, x)
print(f"Время расчёта: {time.time() - curr_time}")

plt.plot(_x, label='x**2+1')
plt.plot(_dydx, label='deriviate')
plt.legend()
plt.show()
print (f"Производная функции x**2 + 2 = {_dydx}")
```
