### Лекция

```
def Sine(A, f, phi, fs, t, N):
    """
    Inputs:
    A  = Амплитуда сигнала
    f  = Частота сигнала  Hz
    phi  = Начальная фаза синусоиды "Радианы"
    fs = Частота семплирования Hz
    t  = Продолжительность сигнала (секунды)
    Output:

    x (numpy array) = 
    """
    T = t/fs  # Период семплирования
    n = np.arange(N) # Кол-во семплов
    x = A*np.sin(2*np.pi*f*n*T + phi)
    return x

```

```
A=1.0       # A  = Амплитуда сигнала
f=10.0      # f  = Частота сигнала  Hz
phi = 1.0   # Начальная фаза синусоиды "Радианы"
fs = 50.0   #  fs = Частота семплирования Hz
t = 0.1     # t  = Продолжительность сигнала (секунды)


plot.plot(Sine(A, f, phi, fs, t, 100))

 

# Подпись графика
plot.title('Sine wave')

 

# Подпись оси
plot.xlabel('Time')
 

plot.grid(True, which='both')
 

plot.axhline(y=0, color='k')
 

plot.show()
 

# Display the sine wave
plot.show()
```

```
# Единичный скачок
# Единичный скачок или функция Хевисайда – вид простых и важных дискретных последовательностей. 
# Он принимает нулевые значения в отрицательной области времени и единичные значения в положительной.

# Математически функция единичного скачка записывается как:

# σ(nT) = δ(nT) + δ(nT - T) + δ(nT – 2T) + ...
n = 6
# time vector
t = np.linspace(-n, n-1, 2*n)

# Delta function
xd = np.zeros(2*n)
xd[n] = 1

# Heaviside function
xh = np.heaviside(t, 1)

# Combine them together
xs = [xh, xd]

# Plot results
fig = plot.figure(figsize=(12, 3), dpi=80)
for i, sig in enumerate(xs):
    plot.subplot(1, 2, i+1)
    plot.stem(t, sig, linefmt='C3', markerfmt='D', use_line_collection=True)
    plot.ylabel('Amplitude')
    plot.xlabel('Samples')
    plot.xticks(t)
    plot.xlim([np.min(t)+1, np.max(t)])
    plot.grid(True)
plot.tight_layout()
```

Модифицировать функцию Sine таким образом, чтобы можно было возвращать составной сигнал - состоящий из комбинаций sin и cos, каждая отдельная гармоническая функция в составном сигнале должна принимать свои параметры:
 A  = Амплитуда сигнала
 f  = Частота сигнала  Hz
 phi  = Начальная фаза синусоиды "Радианы"

 Сделайте вариант с 2-мя и 3-мя гармоническими сигналами


```

def SineCoSine(A1, f1, phi1, A2, f2, phi2, fs, t, N):
    """
    Inputs:
    A  = Амплитуда сигнала
    f  = Частота сигнала  Hz
    phi  = Начальная фаза синусоиды "Радианы"
    fs = Частота семплирования Hz
    t  = Продолжительность сигнала (секунды)
    Output:

    x (numpy array) = 
    """
    T = t/fs  # Период семплирования
    n = np.arange(N) # Кол-во семплов
    x = A1*np.sin(2*np.pi*f1*n*T + phi1) + A2*np.cos(2*np.pi*f2*n*T + phi2)
    return x

A1 = 1.0       # A  = Амплитуда синусоиды
f1 = 10.0      # f  = Частота синусоиды  Hz
phi1 = 1.0     # Начальная фаза синусоиды "Радианы"
A2 = 1.0       # A  = Амплитуда косинусоиды
f2 = 10.0      # f  = Частота косинусоиды  Hz
phi2 = 1.0     # Начальная фаза косинусоиды "Радианы"

fs = 50.0   #  fs = Частота семплирования Hz
t = 0.1     # t  = Продолжительность сигнала (секунды)

plot.plot(SineCoSine(A1, f1, phi1, A2, f2, phi2, fs, t, 100))
```

```

def SineCoSineSine(A1, f1, phi1, A2, f2, phi2, A3, f3, phi3, fs, t, N):
    """
    Inputs:
    A  = Амплитуда сигнала
    f  = Частота сигнала  Hz
    phi  = Начальная фаза синусоиды "Радианы"
    fs = Частота семплирования Hz
    t  = Продолжительность сигнала (секунды)
    Output:

    x (numpy array) = 
    """
    T = t/fs  # Период семплирования
    n = np.arange(N) # Кол-во семплов
    x = (A1*np.sin(2*np.pi*f1*n*T + phi1) + A2*np.cos(2*np.pi*f2*n*T + phi2)) * A3*np.sin(2*np.pi*f3*n*T + phi3)
    return x

A1 = 1.0       # A  = Амплитуда синусоиды
f1 = 10.0      # f  = Частота синусоиды  Hz
phi1 = 1.0     # Начальная фаза синусоиды "Радианы"
A2 = 2.0       # A  = Амплитуда косинусоиды
f2 = 20.0      # f  = Частота косинусоиды  Hz
phi2 = 2.0     # Начальная фаза косинусоиды "Радианы"
A3 = 3.0       # A  = Амплитуда синусоиды
f3 = 30.0      # f  = Частота синусоиды  Hz
phi3 = 3.0     # Начальная фаза синусоиды "Радианы"

fs = 50.0   #  fs = Частота семплирования Hz
t = 0.1     # t  = Продолжительность сигнала (секунды)

plot.plot(SineCoSineSine(A1, f1, phi1, A2, f2, phi2, A3, f3, phi3, fs, t, 100))
```

```
x1 = np.concatenate([np.zeros(8), np.ones(3), np.zeros(8)])     # исходный сигнал
print(f'Исходный сигнал x1 = {x1}')

plot.grid(True)
plot.stem(x1, use_line_collection=True)
plot.show()

# Convolution - свёртка импульса из библиотеки numpy
cv1 = np.convolve(x1, x1, mode='same')
print(f'свёртка импульса из библиотеки numpy = {cv1}')


plot.grid(True)
plot.stem(cv1, use_line_collection=True)
plot.show()

# Convolution - свёртка импульса из лекции
cv2 = convolve(x1,x1)
print(f'свёртка импульса из лекции = {cv2}')

plot.grid(True)
plot.stem(cv2)
plot.show()
```

Самостоятельно ознакомившись с тем, что такое корреляция двух сигналов, вычислить корреляцию между сигналами:
a = np.random.randint(-1, 15, 20)
b = np.random.randint(0, 7, 20)


```

a = np.random.randint(-1, 15, 20)
b = np.random.randint(0, 7, 20)

cfx = np.correlate(a, b, mode='same')
print(cfx)

xl = [a, b, cfx]

plot.figure(figsize=(14, 5))
for i in range(3):
    plot.subplot(3, 1, i+1)
    plot.stem(xl[i], linefmt='C0', markerfmt='D', use_line_collection=True)
    plot.grid(True)
    plot.xlim([-0.5, a.size-0.5])
```

Задача 3

Самостоятельно ознакомившись с тем, что такое Автокорреляционная функция, рассчитать и отобразить графически Автокорреляционную функцию прямоугольного импульса: [np.zeros(8), np.ones(3), np.zeros(8)] , Напишите функцию вычисляющую автокорреляцию.

```
def auto_corr(x):
    res = np.correlate(x, x, mode='same') # np.correlate - вычисление свертки
    return res 

# Signal
x = np.concatenate([np.zeros(8), np.ones(3), np.zeros(8)])

# ACF
cfx = auto_corr(x)

xl = [x, cfx]

# Pot results
plot.figure(figsize=(14, 5), dpi=80)
for i in range(2):
    plot.subplot(2, 1, i+1)
    plot.stem(xl[i], linefmt='C0', markerfmt='D', use_line_collection=True)
    plot.grid(True)
    plot.xlim([-0.5, x.size-0.5])
```
